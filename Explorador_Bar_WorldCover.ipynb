```python
def name_Hmax(_Dpto):
  max_value = None
  max_idx = None
  typeH = None
  perc = (_Dpto/sum(_Dpto))*100

  for idx, num in enumerate(perc):
     if (max_value is None or num > max_value):
          max_value = num
          max_idx = idx
          typeH = _Dpto.name
          nombreDptoMax = nameH[max_idx] + "-"+ typeH

  return nombreDptoMax, max_value
```
