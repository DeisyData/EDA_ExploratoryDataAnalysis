```python
from pylab import *
# importing required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read a csv file
df = pd.read_csv('WorldCover4.csv')
df.info()
```

```python
df.describe()
```

```python
dpto2  = df['LA GUAJIRA']
mylabels = df['NOMBRE_DPT']

dpto  = df['BOLIVAR']
mylabels = df['NOMBRE_DPT']
```

```python
#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]
plt.figure(figsize=(10,10))
#create pie chart
plt.subplot(2,1,1)
plt.pie(dpto, labels = mylabels, colors = colors, autopct='%.0f%%')
plt.title("BOLIVAR", size=18)
plt.subplot(2,1,2)
plt.pie(dpto2, labels = mylabels, colors = colors, autopct='%.0f%%')
plt.title("LA GUAJIRA", size=18)

plt.savefig("DOSGRAFICAS.png")

plt.show()
```
